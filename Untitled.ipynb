```python
def query_group(g) :
    df_p = df[df['label_group']==g] 
    for i in range(df_p.shape[0]) :
        print('label_goup : ', df_p.iloc[i]['label_group'])
        print('title : ', df_p.iloc[i]['title'])
        print('image_phash : ', df_p.iloc[i]['image_phash'])
        im = PIL.Image.open('data/train_images/' + df_p.iloc[i]['image'])
        plt.imshow(im)
        plt.show()

def show_groups(group_list) :
    for g in group_list :
        query_group(g)
        print('.......................................')

def clean(column) :
    to_merge = []
    agg = df.groupby(by=[column])['label_group'].unique()
    agg = agg[(agg.apply(lambda x : len(x)) > 1)]
    for i in range(len(agg)) :
        show_groups(agg.iloc[i])
        merge = input('Merge ?')
        if merge == 'y' :
                to_merge.append(agg.iloc[i])
        print('---------------------------------------------------------------------')
    return to_merge
```
